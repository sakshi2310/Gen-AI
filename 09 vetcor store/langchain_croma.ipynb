```python
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain_core.documents import Document

```

```python
docs1 = Document(
    page_content="virat kohali page data. he is very nice cricketer and his family is very sweet",
    metadata={'team':'royal challengers banglore'}
)

docs2 = Document(
    page_content="rohit sharma is most successful caption In the IPL history leasding the mumbau indian to five titles",
    metadata= {'team':'Mumbai Indians'}
)

docs3 = Document(
    page_content="Ms dhone , famously known as the captopn cool,has lead chennai super kings to the multiple IPL",
    metadata={'team':'chennai super kings'}
)
```

```python
docs = [docs1,docs2,docs3]

vect_store = Chroma(
    embedding_function=OpenAIEmbeddings(),
    persist_directory='croma_db',
    collection_name='sample'
)
```

```python
# add document

vect_store.add_documents(docs)
```

```python
# view document
vect_store.get(include=['embeddings','documents','metadatas'])
```

```python
# similar serach
vect_store.similarity_search(
    query="who is these are bolwer",
    k=1
)
```

```python
vect_store.similarity_search_with_score(
    query="who is bolwer in this",
    k=1
)

# less score is good 
# bec its show ths distance
```

```python
vect_store.similarity_search_with_score(
    query="",
    filter={'team':'cheanni super kings'}
)
```

```python
update_docs1 = Document(
    page_content="new des about virat kohli",
    metadata={'team':'mumbai indians'}
)
vect_store.update_document(document_id="",document=update_docs1)
```
