## method -2 Using the structured tool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class multiplyInput(BaseModel):
    a:int = Field(required=True,description="First number to add")
    b:int = Field(required=True,description="Secound number to add")
```

```python
def mutiply_fucn(a:int,b:int) ->int:
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(func=mutiply_fucn,
                                             name="multiply",
                                             description="multiply the numbers",
                                             args_schema=multiplyInput
                                             )
```

```python
result = multiply_tool.invoke({'a':10,'b':20})
```

```python
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

## Method -3 Using the Base Tool Class

```python
from langchain_core.tools import BaseTool 
from typing import Type
```

```python
class multiplyInput(BaseModel):
    a:int = Field(required=True,description="First number to add")
    b:int = Field(required=True,description="Secound number to add")
```

```python
class MultiplyTool(BaseTool):
    name: str = "Multiply"
    description: str = "Multiply of the two numbers"

    args_schema: Type[BaseModel] = multiplyInput

    def _run(self,a:int,b:int) ->int:
        return a * b
    
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':3,'b':4})
```

```python
print(result)
```
