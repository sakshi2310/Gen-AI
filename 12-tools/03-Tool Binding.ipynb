```python
# tool create 

@tool
def multiply(a:int,b:int) ->int:
    """Give 2 numbers a and b this tool returns their product"""
    return a*b
```

```python
print(multiply.invoke({'a':2,'b':3}))
```

### Tool Binnding

```python
llm = ChatOpenAI()

llm_tools = llm.bind_tools([multiply])
```

```python
llm_tools
```

```python
query = HumanMessage("Can you multiply the 3 with 10")
```

### Tool Excecution

```python
messages = [query]
```

```python
result = llm_tools.invoke(messages)
```

```python
messages.append(result)
```

```python
result.tool_calls[0]
```

```python
result.tool_calls[0]['args'] # to give the result
```

```python
tool_reuslt = multiply.invoke(result.tool_calls[0])
```

```python
messages.append(tool_reuslt)
```

```python
llm.invoke(messages)
```
