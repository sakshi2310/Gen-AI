```python
from langchain_core.tools import tool

# custome tools 

@tool
def mutiple(a:int,b:int) -> int:
    """Multiply two numbers"""
    return a*b

# step -3 Add tool decorator

@tool
def add(a:int,b:int) -> int:
    """add two numbers"""
    return a+b
```

```python
class Mathtoolkit:
    def get_tool(self):
        return[add,mutiple]
```

```python
toolkit = Mathtoolkit()
tools = toolkit.get_tool()

for tool in tools:
    print(tool.name,"==>",tool.description)
```
