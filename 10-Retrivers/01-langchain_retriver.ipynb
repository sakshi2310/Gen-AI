## Source based retriver - wikipedia

```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
# retriver 
retriver = WikipediaRetriever(top_k_results=2,lang='en')
```

```python
query = "give the history of the india"

docs = retriver.invoke(query)
```

```python
docs
```

```python
for i,docs in enumerate(docs):
    print(f"\n---------result {i+1}------")
    print(f"content : \n{docs.page_content}-----")
```

## Vectore store retrivers

```python
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings
from langchain_core.documents import Document
```

```python
document = [
    Document(page_content="Langchain helps develpoers build LLM applications easilty"),
    Document(page_content="Chorma is a vector database optimized for the LLm based search"),
    Document(page_content="embedding is convert the text into high-dimentional vectors")
]
```

```python
# 
embedding_model= OpenAIEmbeddings()

vectorstore = Chroma.from_documents(
    documents=document,
    embedding=embedding_model,
    collection_name="my_collection"
)

retriver = vectorstore.as_retriever(search_kwargs={'k':2})

query  = "what is chroma used for.."
result  = retriver.invoke(query)
```
