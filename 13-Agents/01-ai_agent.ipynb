```python
import os
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
import requests
from langchain_community.tools import DuckDuckGoSearchRun
```

```python
search_tool = DuckDuckGoSearchRun()
result = search_tool.invoke("top news in india today")
result
```

```python
llm = ChatOpenAI()
llm.invoke('hi')
```

```python
from langchain.agents.react.agent import create_react_agent
from langchain.agents import AgentExecutor
from langchain import hub
# Not working
```

```python
prmopt = hub.pull("hwchase17/react") ## pulls the standard react agent prompt
```

```python
# step - 3 creat the react agetn manually with the pulled prompt
agent = create_react_agent(
    llm = llm,
    tools=[search_tool],
    prompt = prmopt
)
```

```python
## step - 4 wrap it in the agent agent Executor
agent_exe = AgentExecutor(
    agent = agent,
    tools = [search_tool],
    verbose = True
)
```

```python
response = agent_exe.invoke({'input':'3 ways to reach goa from the delhi'})
```
